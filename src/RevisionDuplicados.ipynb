```python
# Determinar cual de las zonas es el "root" para asi hacer clustering
from fuzzywuzzy import fuzz
#for each root identify which zone name contains it:
# for r in roots:
#     print
G = "ROOT_NODE_EMPTY"
town_roots = []
for x in single_town_zones_X:
    for y in single_town_zones_X:
        r1, r2= fuzz.ratio(x, y), fuzz.partial_ratio(x, y) 
        if r1==r2==100 and len(x)==len(y) and G not in x:
           # print(G, x, len(G), len(x))
            G = x
            town_roots.append(G)
            
        if r1>90 and r2>90:
            print(f"{{{G}}} >>> {{{ x }}} --> {{{ y }}}", fuzz.ratio(x, y), fuzz.partial_ratio(x, y) )
        
```
